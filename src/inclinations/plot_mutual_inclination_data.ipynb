```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
SHOW_PLOT = True
SAVE_PLOT = True
```

```python
def plot_histogram():
    N_BINS = 20000

    filepath = os.path.join(os.pardir, os.pardir, 'results', 'inclinations', 'mean_median_inclination_variations.csv')
    data = pd.read_csv(filepath)

    variation_data = data['Variation'].astype(float)
    frac_variation_data = data['Fractional Variation'].astype(float)

    (fig, axes) = plt.subplots(1, 2)
    fig.set_size_inches(18, 9)

    plt.subplot(1, 2, 1)

    ax1 = axes[0]
    ax1.minorticks_on()

    ax1.grid(True, axis = 'x', alpha = 0.25)
    ax1.hist(frac_variation_data, bins = N_BINS)
    ax1.set_xlim(-0.4, 0.8)

    ax1.xaxis.set_major_locator(tck.LinearLocator(14))
    ax1.xaxis.set_major_formatter(tck.FuncFormatter(lambda x,pos: '%.2f' % x))

    ax1.set_xlabel('Fractional Variation $(mean - median)/median$', fontsize = 15)
    ax1.set_ylabel('Frequency', fontsize = 15)

    plt.subplot(1, 2, 2)

    ax2 = axes[1]
    ax2.minorticks_on()

    ax2.grid(True, axis = 'x', alpha = 0.25)
    ax2.hist(variation_data, bins = N_BINS)
    ax2.set_xlim(-0.015, 0.015)

    ax2.xaxis.set_major_locator(tck.LinearLocator(14))
    ax2.xaxis.set_major_formatter(tck.FuncFormatter(lambda x,pos: '%.3f' % x))
    ax2.set_xlabel('Variation Degrees ($mean - median$)', fontsize = 15)
    ax2.set_ylabel('Frequency', fontsize = 15)

    plt.subplots_adjust(wspace = 0.15)
    plt.suptitle('Distribution of Mutual Inclination Mean-Median Variation', fontsize = 30, y = 0.95)

    if SHOW_PLOT:
        plt.show()

    if SAVE_PLOT:
        savepath = os.path.join(os.pardir, os.pardir, 'results', 'inclination_histogram')
        fig.savefig(savepath, bbox_inches = 'tight', dpi = 300)
```

```python
plot_histogram()
```
